<h1> Working with a data set of fake names from the UK with various details included. The dataset is missing data and we need to see visually where it is so <h1>

```python
#The package Missingno makes it easy to visualize missing data in any data set
pip install missingno
```

```python
import missingno as msno
%matplotlib inline
```

```python
import pandas as pd
```

```python
#load dataset into a dataframe. We just need the dataset to open, errors about bad lines dont concern us. 
FNuk = pd.read_csv(r'C:\Users\Daniel\Data Science Projects\20190628 UK Fake Names\1. Original Data\FakeNamesUK.csv', error_bad_lines = False, encoding = 'latin1')
```

```python
#get an idea of how the data looks like
FNuk.head(5)
```

```python
# Need length of dataset to match sample paramater in msno.matrix(data_set.sample(paramater)). If greater than, wont work. If less than, incomplete picture
len(FNuk)
```

```python
# visualize all instances of missing data. White lines across gray bar indicate missing rows of data
msno.matrix(FNuk.sample(99991))
```

```python
# visualize the number of complete rows per column
msno.bar(FNuk.sample(99991))
```

```python
#visualize and find correlation for missing data between pairs of categories
msno.heatmap(FNuk)
```

```python
#find and visualize non-linear hierarchical clusters 
msno.dendrogram(FNuk)
```
